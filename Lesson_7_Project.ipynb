```python
# Import Libraries
from PIL import Image
from sklearn.cluster import KMeans
```

```python
#Open the image “dog.jpeg.”
image = Image.open('dogs.jpeg')
```

```python
# Display the image
from matplotlib import pyplot as plot

%matplotlib inline

plot.imshow(image)
```

```python
# Convert the image to array
import numpy as np

image_np = np.asarray(image)
image_np
```

```python
# Find the shape of image
image_np.shape
```

```python
# Convert into 2d array
image_np = image_np.reshape((185*272,3))
```

```python
image_np.shape
image_np
```

```python
#Use K-means clustering with k set to 3 and cluster the image.
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(image_np)
```

```python
# Predict the cluster label of every pixel in the image 
labels = kmeans.predict(image_np)
labels
```

```python
labels = labels.reshape((185,272))
```

```python
# plot it back as an image.
plot.imshow(labels)
```

```python
# Find out the three dominant colors in the image.
# [Hint: The cluster centers should correspond to three dominant colors]
centers = [centers.astype(int) for centers in kmeans.cluster_centers_]
centers
```

```python
blob = np.ones((100,100,3))
blob
```

```python
blob1 = blob * centers[0]
plot.imshow((blob1).astype(np.uint8))
```

```python
blob2 = blob * centers[1]
plot.imshow((blob2).astype(np.uint8))
```

How to find the k in K meeans clustering? We use elbow plot for this. The below code calculates sum of squared errors (SSE) for k values 1 to 10. We plot it in a plot. The best k value is the one that is found in the elbow of the plot. From the below plot, we find that the best k value is 3.

```python
k_rng = range(1,10)
sse = []
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(image_np)
    sse.append(km.inertia_)
```

```python
sse
```

```python
plot.xlabel('K')
plot.ylabel('Sum of Squared Errors(SSE)')
plot.plot(k_rng,sse)
```
